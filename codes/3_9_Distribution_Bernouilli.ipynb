## Distribution binomiale

Nous pouvons simuler le processus de Bernoulli avec des cas générés aléatoirement et
compter le nombre de réussites sur le nombre d'épreuve donné. Ceci peut être réalisé via la
fonction $NumPy binomial()$. Cette fonction prend le nombre total d'épreuves et la probabilité de réussite comme arguments et renvoie le nombre d'issues positives sur l'ensemble des épreuves pour une simulation.

En probabilité, une épreuve de Bernoulli de paramètre p (réel compris entre 0 et 1) est une expérience aléatoire (c'est-à-dire soumise au hasard) comportant deux issues, le succès ou l'échec. 

Exemple de simulation d'un processus binomial et de comptage des succès

```python
from numpy.random import binomial
```

Définir les paramètres de la distribution

La probabilité de succès donné pour chaque essais $p$ est

```python
p = 0.3
```

Nombre total d'épreuves est $k$ = 100

```python
k=100
```

Exécuter une seule simulation

```python
success = binomial (k,p)
```

```python
print('Total success: %d' %success)
```
