## Distribution binomiale

Nous pouvons calculer les moments de cette distribution, plus précisément la valeur attendue ou moyenne et la variance à l'aide de la fonction $SciPy\ binom.stats()$. 

```python
from scipy.stats import binom
```

Définir les paramètres de la distribution

La probabilité de succès donné pour chaque essais $p$ est

```python
p = 0.3
```

Nombre total d'épreuves est $k$ = 100

```python
k=100
```

Calculer les moments de distribution binomiale

```python
mean,var,_,_ = binom.stats(k, p, moments='mvsk')
```

```python
print('Mean=%.3f, Variance=%.3f' %(mean, var))
```
