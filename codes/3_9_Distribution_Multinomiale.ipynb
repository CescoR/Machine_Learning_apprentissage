## Distribution multinomiale

Nous pouvons utiliser la fonction $NumPy\ multinomial()$ pour simuler 100 essais indépendants et résumer le nombre de fois où l'événement a abouti à chacune des catégories données. La fonction prend à la fois le nombre d'essais et les probabilités pour chaque catégorie sous forme de liste. 

```python
from numpy.random import multinomial
```

Définir les paramètres de la distribution.
Les probabilités $p$ pour chaque catégorie sous forme de liste.

```python
p = [1.0/3.0, 1.0/3.0, 1.0/3.0]
```

Nombre total d'essais est $k$ = 100

```python
k=100
```

Calculer les moments de distribution multinomiale.

```python
cases = multinomial(k, p)
```

Impression du nombre d'événements pas catagérorie

```python
for i in range(len(cases)):
    print('Case %d: %d' % (i+1, cases[i]))
```
